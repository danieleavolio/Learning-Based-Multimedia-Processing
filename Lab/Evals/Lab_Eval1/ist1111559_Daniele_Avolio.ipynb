### 1. Image Manipulation

```python
import cv2
import matplotlib.pyplot as plt
```

```python
i03 = cv2.imread("images/i03.jpg")

rgb_img = cv2.cvtColor(i03, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
plt.show()

```

1.1) Grayscale Image

```python
gray_img = cv2.cvtColor(i03, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img, cmap='gray')
plt.show()
```

1.2) Upside-down image

```python
reversed = rgb_img[::-1]


plt.imshow(reversed, cmap='gray')
plt.show()
```

1.3) Spatially Subsample

```python
factors = [4,16,64]
real_images = [rgb_img, gray_img, reversed]

factored_images = []

for i in factors:
    factored_images.append(rgb_img[::i, ::i])
```

1.4) Subplot plotting

```python
# I created this method to use it in for loop
def show_img(img):
    if len(img.shape) == 3:
        plt.imshow(img)
    else:
        plt.imshow(img, cmap='gray')
```

```python
plt.figure()
for i in range(3):
    plt.subplot(2, 3, i+1)
    show_img(real_images[i])
    plt.axis('off')
for i in range(3):
    plt.subplot(2, 3, i+4)
    show_img(factored_images[i])
    plt.axis('off')

plt.show()
```

### Discussion:
A problem I have is to get the image reversed but not mirrored.

A tricky thing was to get the spatial subsample. Basically it's only getting the images with a step based on the factor we wanted to use.

The only thing I didn't like was the subplot with the for loop. To solve this, it's easy to create a object with title and image and then iterate over it. I didn't do this for time reasons.
